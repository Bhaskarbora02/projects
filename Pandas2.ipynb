```python
#TYPES OF VARIABLES

# THERE ARE 2 TYPES OF VARIABLES

# 1. QUANTITATIVE/NUMERICAL
# 2. QUALITATIVE/CATEGORICAL


# QUANTITATIVE

# 1. DISCRETE
#IT WILL ALWAYS BE AN INTEGER 
#will always be countable parameters
#population,no.of cars on road

# 2. CONTINOUS
#IT MAY OR MAY NOT BE INTEGER
#will always be measureable parameters
#volume,#densite#height/weight

# QUALITATIVE

# 1.BINOMIAL
##when we have 2 outcomes it will be binomial
#T/F,W/L,H/T,S/DS,Y/N

# # 2.NOMINAL
# WHEN WE HAVE MORE THAN 2 OUTCOMES IT WILL BE NOMINAL
# IT MAY OR MAY NOT BE IN ORDER 
# IF WE CHANGE THE ORDER NO EFFECT ON MEANINGS
# #NAME,APPAREL  CATEGORY

# # 3.ODINAL
# WHEN WE HAVE MORE THAN 2 OUTCOMES IT WILL BE ODINAL
# BUT IT NEEDS TO BE IN ORDER
# IF WE CHANGE THE ORDER THE MEANINGS CHANGES
# ##MOVIE RATINGS,RANKS OF STUDENTS,SIZE OF CLOTHES
```
